## Juego

#### **Clases**
- 1 clase denominada ``Dron``:
    - representa un Dron en un juego de mesa.
    - consta de dos parámetros que lo identifican: nombre y pisición.
    - se inicializa en ``0,0`` con nombres ``Timon`` y ``Sid``

- 1 clase denominada ``Tablero``:
    - que representa un tablero de juego de mesa.
    - tiene tamaño que se representa por: x e y.
    - se inicializa en tamaño ``50,50``

#### **Comportamiento**
- El dron se mueve en 4 direcciones: arriba, abajo, izquierda, derecha
- En el trablero, el dron puede lanzar bombas

#### **Jugada**
- Una jugada representa el juego en marcha.
- Primero se mueven los drones en diferentes direcciones
- Luego, cada dron lanza una bomba
- Si la posición de la bomba lanzada por un dron coincide con la posicion del rival, entonces el dron gana.

#### 1 - Clase Dron

- Cree la clase Dron
- Implemente la clase.
    - ``Dron``
- Implemente los métodos
    - ``mover_arriba``, ``mover_abajo``, ``mover_izquierda``, ``mover_derecha``, y  ``lanzar_bomba``.

```python
class Dron:
    def __init__(self, nombre="Dron", x=0, y=0):
        self.nombre = nombre
        self.posicion = (x, y)

    def mover_arriba(self):
        x, y = self.posicion
        if y < 50:
            self.posicion = (x, y + 1)

    def mover_abajo(self):
        x, y = self.posicion
        if y > 0:
            self.posicion = (x, y - 1)

    def mover_izquierda(self):
        x, y = self.posicion
        if x > 0:
            self.posicion = (x - 1, y)

    def mover_derecha(self):
        x, y = self.posicion
        if x < 50:
            self.posicion = (x + 1, y)

    def lanzar_bomba(self, otro_dron): #es explicado mejor en el apartado de jugada, el lanzamiento de la bomba es random
        print(f"{self.nombre} ha lanzado una bomba a la posición {otro_dron.posicion} de {otro_dron.nombre}.")
```

```python
import unittest

class PruebaDron(unittest.TestCase):
    def prueba_posicion_inicial(self):
        dron = Dron()
        posicion_esperada = (0, 0)
        self.assertEqual(dron.posicion, posicion_esperada)
        print(f"Posición inicial: {dron.posicion}. Se esperaba: {posicion_esperada}")
    
    def prueba_mover_arriba(self):
        dron = Dron()
        dron.mover_arriba()
        posicion_esperada = (0, 1)
        self.assertEqual(dron.posicion, posicion_esperada)
        print(f"Desplazamiento hacia arriba: {dron.posicion}. Se esperaba: {posicion_esperada}")

    def prueba_mover_abajo(self):
        dron = Dron()
        dron.mover_abajo()
        posicion_esperada = (0, 0)
        self.assertEqual(dron.posicion, posicion_esperada)
        print(f"Desplazamiento hacia abajo: {dron.posicion}. Se esperaba: {posicion_esperada}")
    
    def prueba_mover_izquierda(self):
        dron = Dron()
        dron.mover_izquierda()
        posicion_esperada = (0, 0)
        self.assertEqual(dron.posicion, posicion_esperada)
        print(f"Desplazamiento hacia la izquierda: {dron.posicion}. Se esperaba: {posicion_esperada}")

    def prueba_mover_derecha(self):
        dron = Dron()
        dron.mover_derecha()
        posicion_esperada = (1, 0)
        self.assertEqual(dron.posicion, posicion_esperada)
        print(f"Desplazamiento hacia la derecha: {dron.posicion}. Se esperaba: {posicion_esperada}")

if __name__ == '__main__':
    unittest.main(argv=['primer-argumento-ignorado'], exit=False)
```

#### 3 - Jugada

- Implemente una jugada (con o sin clase)
- Mueva los drones en cierta dirección.
- Debe atacar lanzando bombas.
- Uno de los drones debe ganar y hacer un anuncio al respecto.

```python
import random
# Crear instancias de los drones Timon y Sid con posiciones iniciales
timon = Dron(nombre="Timon", x=2, y=25)
sid = Dron(nombre="Sid", x=10, y=43)

# Iniciar el juego
while True:
    # Turno de Timon
    movimiento = random.choice([timon.mover_arriba, timon.mover_abajo, timon.mover_izquierda, timon.mover_derecha])
    movimiento()
    timon.lanzar_bomba(sid)
    
    # Verificar si Timon ha ganado
    if timon.posicion == sid.posicion:
        print(f"{timon.nombre} ha ganado el juego.")
        break

    # Turno de Sid
    movimiento = random.choice([sid.mover_arriba, sid.mover_abajo, sid.mover_izquierda, sid.mover_derecha])
    movimiento()
    sid.lanzar_bomba(timon)

    # Verificar si Sid ha ganado
    if sid.posicion == timon.posicion:
        print(f"{sid.nombre} ha ganado el juego.")
        break
```

Prueba de jugada

```python
import random
import unittest

class TestJuegoDrones(unittest.TestCase):
    def test_game_result(self):
        timon = Dron(nombre="Timon", x=2, y=25)
        sid = Dron(nombre="Sid", x=10, y=43)
        while True:
            movimiento = random.choice([timon.mover_arriba, timon.mover_abajo, timon.mover_izquierda, timon.mover_derecha])
            movimiento()
            timon.lanzar_bomba(sid)
            if timon.posicion == sid.posicion:
                self.assertEqual(f"{timon.nombre} ha ganado el juego.", "Timon ha ganado el juego.")
                break

            movimiento = random.choice([sid.mover_arriba, sid.mover_abajo, sid.mover_izquierda, sid.mover_derecha])
            movimiento()
            sid.lanzar_bomba(timon)
            if sid.posicion == timon.posicion:
                self.assertEqual(f"{sid.nombre} ha ganado el juego.", "Sid ha ganado el juego.")
                break

if __name__ == '__main__':
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
